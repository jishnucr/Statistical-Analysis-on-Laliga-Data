Convert 'Debut' column to numeric and filter teams

Filtering Teams by Start year(1930 - 1980)

```python


filtered_teams = df[(df['Debut'] >= 1930) & (df['Debut'] <= 1980)]
print(filtered_teams[['Team', 'Debut']])
```

Top 5 teams Based on the points

```python
df['Points'] = pd.to_numeric(df['Points'])
top_5_teams = df.nlargest(5, 'Points')
print(top_5_teams[['Team', 'Points']])
```

Goal Difference Function

```python
def Goal_diff_count(row):
    return int(row['GoalsFor']) - int(row['GoalsAgainst'])


df['GoalDiff'] = df.apply(Goal_diff_count, axis=1)


max_goal_diff_team = df.loc[df['GoalDiff'].idxmax()]

min_goal_diff_team = df.loc[df['GoalDiff'].idxmin()]

print(f"Team with maximum goal difference: {max_goal_diff_team['Team']}")
print(f"Team with minimum goal difference: {min_goal_diff_team['Team']}")
```

Winning Percentage

```python

df['GamesWon'] = pd.to_numeric(df['GamesWon'])
df['GamesPlayed'] = pd.to_numeric(df['GamesPlayed'])


df['WinningPercent'] = (df['GamesWon'] / df['GamesPlayed']) * 100
df['WinningPercent'].replace([np.inf, -np.inf, np.nan], 0, inplace=True)


top_winning_teams = df.nlargest(5, 'WinningPercent')
print(top_winning_teams[['Team', 'WinningPercent']])
```

Group by Best Position and Sum Points

```python

df['BestPosition'] = pd.to_numeric(df['BestPosition'])


grouped_data = df.groupby('BestPosition')['Points'].sum().reset_index()
print(grouped_data)
```
